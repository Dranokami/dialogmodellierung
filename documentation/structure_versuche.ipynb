## MENSAAPP

here different structure suggestions are visualised

```python
# -*- coding: utf-8 -*-
from graphviz import Digraph


dot = Digraph(comment='MensaApp')
dot  #doctest: +ELLIPSIS
```

```python
dot.node('A', 'MensaApp')
dot.node('B', '[Menue 1] und [Menue 2] \n[anderes Menue]^[Preise]^[anderer Tag]')
dot.node('D','OR')
dot.node('E','Info [keyword]')
dot.node('F','Frage: [keyword 1] v [keyword 2]')
dot.node('G','FINAL')



dot.edges(['AB','BD','BE','EF', 'FE','BG','EG','FG', 'ED', 'FD', 'DB'])
print dot
```

```python
dot.render('test-output/mensaapp.gv', view=True)
```

## Finde Datum in Mensabaum

```python
import re
```

```python
text = "hallo liebe welt. heute ist der 06.09.2017"
```

```python
compiler = re.compile(r"[0-3][0-9]\.[0-1][0-9]\.20[0-9][0-9]")
found = compiler.search(text)
found.group()
```
